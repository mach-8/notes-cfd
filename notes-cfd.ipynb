# PART 1: GENERAL GOVERNING EQUATIONS
***

**Continuity Equation**
    
* Mass conservation requires:

    $$\text{Mass Accumulation + [Mass Flow Out - Mass Flow In]} = 0$$

    $$\frac{\left(\rho|_{t+\Delta t} - \rho|_{t}\right)\Delta x \Delta y \Delta z}{\Delta t} + \left[\rho u|_{x+\Delta x} - \rho u|_x\right]\Delta y \Delta z + \left[\rho u|_{y+\Delta y} - \rho u|_y\right]\Delta x \Delta z + \left[\rho u|_{z+\Delta z} - \rho u|_z\right]\Delta x \Delta y = 0$$
    
* Dividing by the volume of the fluid element, $\Delta x \Delta y \Delta z$:

    $$\frac{\left[\rho|_{t+\Delta t} - \rho|_{t}\right]}{\Delta t} + \frac{\left[\rho u|_{x + \Delta x} - \rho u|_x\right]}{\Delta x} + \frac{\left[\rho u|_{y + \Delta y} - \rho u|_y\right]}{\Delta y} + \frac{\left[\rho u|_{z + \Delta z} - \rho u|_z\right]}{\Delta z} = 0$$

* In the limits of $\Delta t$, $\Delta x$, $\Delta y$, and $\Delta z$ approaching zero:

    $$\frac{\partial \rho}{\partial t} + \frac{\partial\left(\rho u_x\right)}{\partial x} + \frac{\partial\left(\rho u_y\right)}{\partial y}  + \frac{\partial\left(\rho u_z\right)}{\partial z} = 0$$

**Momentum Equations**

* The rate of change of a conserved quantity $\phi(t,x,y,z)$ following a fluid particle in motion is given by the total derivative of the quantity:

    $$\frac{D\phi}{Dt} = \frac{\partial \phi}{\partial t} + \frac{\partial \phi}{\partial x} \frac{\partial x}{\partial t} + \frac{\partial \phi}{\partial y} \frac{\partial y}{\partial t} + \frac{\partial \phi}{\partial z} \frac{\partial z}{\partial t}$$

* Recognizing that $\frac{\partial x}{\partial t} = u_x$, $\frac{\partial y}{\partial t} = u_y$, and $\frac{\partial z}{\partial t} = u_z$:

    $$\frac{D\phi}{Dt} = \frac{\partial \phi}{\partial t} + u_x\frac{\partial \phi}{\partial x} + u_y\frac{\partial \phi}{\partial y} + u_z\frac{\partial \phi}{\partial z}$$

* The above equation describes the rate of change of a conserved quantity per unit mass using a Lagrangian approach--the rate of change of the conserved quantity per unit volume (Eulerian approach) can be obtained by multiplying the prior equation by the density:

    $$\rho\frac{D\phi}{Dt} = \rho\left[\frac{\partial \phi}{\partial t} + u_x\frac{\partial \phi}{\partial x} + u_y\frac{\partial \phi}{\partial y} + u_z\frac{\partial \phi}{\partial z}\right]$$

* The above equation can alternatively be derived considering a fluid element fixed in space. The net rate of change of a conserved quantity for the fluid element comprises of time rate of change and advection from the surroundings through the boundaries of the fluid element:

    $$\text{Net rate of change = time rate of change + advection from the surroundings}$$
    
    $$\rho \Delta x \Delta y \Delta z \frac{D \phi}{D t} = \frac{\left[\rho \phi|_{t+\Delta t} - \rho \phi|_t\right]\Delta x \Delta y \Delta z}{\Delta t} + [\rho u_x \phi|_{x+\Delta x} - \rho u_x \phi|_x]\Delta y \Delta z + [\rho u_y \phi|_{y+\Delta y} - \rho u_y \phi|_y] \Delta x \Delta z  + [\rho u_z \phi|_{z+\Delta z} - \rho u_z \phi|_z]\Delta x \Delta y$$

    $$\rho \frac{D \phi}{D t} = \frac{\left[\rho \phi|_{t+\Delta t} - \rho \phi|_t\right]}{\Delta t} + \frac{[\rho u_x \phi|_{x+\Delta x} - \rho u_x \phi|_x]}{\Delta x} + \frac{[\rho u_y \phi|_{y+\Delta y} - \rho u_y \phi|_y]}{\Delta y}  + \frac{[\rho u_z \phi|_{z+\Delta z} - \rho u_z \phi|_z]}{\Delta z}$$

    $$\rho \frac{D \phi}{D t} = \frac{\partial \left(\rho \phi\right)}{\partial t} + \frac{\partial \left(\rho u_x \phi\right)}{\partial x} + \frac{\partial \left(\rho u_y \phi\right)}{\partial y}  + \frac{\partial \left(\rho u_z \phi\right)}{\partial z}$$

    $$\rho \frac{D \phi}{D t} = \rho\left[\frac{\partial \phi}{\partial t} + u_x \frac{\partial \phi}{\partial x} + u_y\frac{\partial \phi}{\partial y}  + u_z\frac{\partial \phi}{\partial z}\right] + \phi \cancel{\left[\frac{\partial \rho}{\partial t} + \frac{\partial \left(\rho u_x\right)}{\partial x} + \frac{\partial \left(\rho u_y\right)}{\partial y} + \frac{\partial \left(\rho u_z\right)}{\partial z}\right]}$$

    $$\rho \frac{D \phi}{D t} = \rho\left[\frac{\partial \phi}{\partial t} + u_x \frac{\partial \phi}{\partial x} + u_y\frac{\partial \phi}{\partial y}  + u_z\frac{\partial \phi}{\partial z}\right]$$

* From Newton's second law of motion, the rate of change of momentum (conserved quantity) is equal to the sum of forces acting on the fluid element:
    
    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = \sum F_x$$
    
    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = \sum F_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = \sum F_z$$

* Forces (per unit volume) under consideration:
    
    1. Pressure forces: $-\frac{\partial p}{\partial x}$, $-\frac{\partial p}{\partial y}$, $-\frac{\partial p}{\partial z}$
    
    2. Viscous shear forces: $\frac{\partial \tau_{xx}}{\partial x}$, $\frac{\partial \tau_{yx}}{\partial x}$, $\frac{\partial \tau_{zx}}{\partial x}$, $\frac{\partial \tau_{xy}}{\partial y}$, $\frac{\partial \tau_{yy}}{\partial y}$, $\frac{\partial \tau_{zy}}{\partial y}$, $\frac{\partial \tau_{xz}}{\partial z}$, $\frac{\partial \tau_{yz}}{\partial z}$, $\frac{\partial \tau_{zz}}{\partial z}$

    3. Gravity: $\rho g_x$, $\rho g_y$, $\rho g_z$

    4. Other forces: $S_x$, $S_y$, $S_z$

* Subbing in the forces above:

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \frac{\partial \tau_{xx}}{\partial x} + \frac{\partial \tau_{yx}}{\partial x} + \frac{\partial \tau_{zx}}{\partial x} + \rho g_x + S_x$$
    
    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \frac{\partial \tau_{xy}}{\partial y} + \frac{\partial \tau_{yy}}{\partial y} + \frac{\partial \tau_{zy}}{\partial y} + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \frac{\partial \tau_{xz}}{\partial z} + \frac{\partial \tau_{yz}}{\partial z} + \frac{\partial \tau_{zz}}{\partial z} + \rho g_z + S_z$$

# PART 2: GOVERNING EQUATIONS FOR NEWTONIAN FLUIDS
***

**Continuity Equation**

* From part 1:

    $$\frac{\partial \rho}{\partial t} + \frac{\partial\left(\rho u_x\right)}{\partial x} + \frac{\partial\left(\rho u_y\right)}{\partial y}  + \frac{\partial\left(\rho u_z\right)}{\partial z} = 0$$

**Momentum Equations** 

* For Newtonian fluids, we can use stoke's relations for viscous stresses:

    linear transformations: $\tau_{xx} = 2\mu \frac{\partial u_x}{\partial x} + \lambda \nabla \cdot \mathbf{u}$, $\tau_{yy} = 2\mu \frac{\partial u_y}{\partial y} + \lambda \nabla \cdot \mathbf{u}$, $\tau_{zz} = 2\mu \frac{\partial u_z}{\partial z} + \lambda \nabla \cdot \mathbf{u}$

    volumetric transformations: $\tau_{xy} = \tau_{yx} = \mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)$, $\tau_{xz} = \tau_{zx} = \mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)$, $\tau_{yz} = \tau_{zy} = \mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)$

* Subbing in Stokes' relations:
    
    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \frac{\partial}{\partial x}\left[2\mu \frac{\partial u_x}{\partial x} + \lambda \nabla \cdot \mathbf{u} \right] + \frac{\partial}{\partial y}\left[\mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)\right]  + \frac{\partial}{\partial z}\left[\mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\right] + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \frac{\partial}{\partial x}\left[\mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)\right] + \frac{\partial}{\partial y}\left[2\mu \frac{\partial u_y}{\partial y} + \lambda \nabla \cdot \mathbf{u} \right] + \frac{\partial}{\partial z}\left[\mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\right] + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \frac{\partial}{\partial x}\left[\mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\right] + \frac{\partial}{\partial y}\left[\mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\right] + \frac{\partial}{\partial z}\left[2\mu \frac{\partial u_z}{\partial z} + \lambda \nabla \cdot \mathbf{u} \right] + \rho g_z + S_z$$

* Re-arranging:

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \mu \frac{\partial}{\partial x}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \mu \frac{\partial}{\partial y}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial y}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \mu \frac{\partial}{\partial z}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_z + S_z$$

* In more compact form:

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \mu \frac{\partial}{\partial x}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \mu \frac{\partial}{\partial y}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial y}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \mu \frac{\partial}{\partial z}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_z + S_z$$

# PART 3: GOVERNING EQUATIONS FOR INCOMPRESSIBLE FLOWS
***

**Continuity Equation**

* From part 2:
    
    $$\frac{\partial \rho}{\partial t} + \frac{\partial\left(\rho u_x\right)}{\partial x} + \frac{\partial\left(\rho u_y\right)}{\partial y}  + \frac{\partial\left(\rho u_z\right)}{\partial z} = 0$$

* For incompressible flows, density is constant $\left(\frac{\partial \rho}{\partial t} = 0\right)$:

    $$\frac{\partial u_x}{\partial x} + \frac{\partial u_y}{\partial y}  + \frac{\partial u_z}{\partial z} = 0$$

* In more compact form:

    $$\nabla \cdot \mathbf{u} = 0$$

**Momentum Equations**

* From part 2:
    
    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \mu \frac{\partial}{\partial x}\left(\cancel{\nabla \cdot \mathbf{u}}\right)  +\frac{\partial}{\partial x}\left(\lambda \cancel{\nabla \cdot \mathbf{u}}\right) + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \mu \frac{\partial}{\partial y}\left(\cancel{\nabla \cdot \mathbf{u}}\right)  + \frac{\partial}{\partial y}\left(\lambda \cancel{\nabla \cdot \mathbf{u}}\right) + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \mu \frac{\partial}{\partial z}\left(\cancel{\nabla \cdot \mathbf{u}}\right)  + \frac{\partial}{\partial x}\left(\lambda \cancel{\nabla \cdot \mathbf{u}}\right) + \rho g_z + S_z$$

* From continuity, $\nabla \cdot \mathbf{u} = 0$: 

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \rho g_z + S_z$$

# PART 4: GOVERNING EQUATIONS FOR 2-D FLOWS
***

**Continuity Equation**

* From part 3, we only keep x and y terms:

    $$\frac{\partial u_x}{\partial x} + \frac{\partial u_y}{\partial y} = 0$$

**Momentum Equations**

* From part 3, we keep x- and y-components of the momentum equations and set $u_z = 0$:

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial^2 u_x}{\partial x^2} + \frac{\partial^2 u_x}{\partial y^2}\right) + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} \right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial^2 u_y}{\partial x^2} + \frac{\partial^2 u_y}{\partial y^2}\right) + \rho g_y + S_y$$

# PART 5: VORTICITY - STREAM FUNCTION FORMULATION
***

**Continuity Equation**

* From part 4:

    $$\frac{\partial u_x}{\partial x} + \frac{\partial u_y}{\partial y} = 0$$

**Momentum Equations**

* If we set $g_x=g_y=0$ and $S_x = S_y = 0$:

    $$\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} = -\frac{1}{\rho}\frac{\partial p}{\partial x} + \nu\left(\frac{\partial^2 u_x}{\partial x^2} + \frac{\partial^2 u_x}{\partial y^2} \right)$$

    $$\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} = -\frac{1}{\rho}\frac{\partial p}{\partial y} + \nu\left(\frac{\partial^2 u_y}{\partial x^2} + \frac{\partial^2 u_y}{\partial y^2} \right)$$

* Differentiating the x-component and y-component of momentum equations with respect to y and x, respectively:

    $$\frac{\partial}{\partial y}\left(\frac{\partial u_x}{\partial t}\right) + \frac{\partial}{\partial y}\left(u_x\frac{\partial u_x}{\partial x}\right) + \frac{\partial}{\partial y}\left(u_y\frac{\partial u_x}{\partial y}\right) = -\frac{\partial}{\partial y}\left(\frac{1}{\rho}\frac{\partial p}{\partial x}\right) + \nu\frac{\partial}{\partial y}\left[\left(\frac{\partial^2 u_x}{\partial x^2} + \frac{\partial^2 u_x}{\partial y^2} \right)\right]$$

    $$\frac{\partial}{\partial x}\left(\frac{\partial u_y}{\partial t}\right) + \frac{\partial}{\partial x}\left(u_x\frac{\partial u_y}{\partial x}\right) + \frac{\partial}{\partial x}\left(u_y\frac{\partial u_y}{\partial y}\right) = -\frac{\partial}{\partial x}\left(\frac{1}{\rho}\frac{\partial p}{\partial y}\right) + \nu\frac{\partial}{\partial x}\left[\left(\frac{\partial^2 u_y}{\partial x^2} + \frac{\partial^2 u_y}{\partial y^2} \right)\right]$$

* Re-arranging above equations:

    $$\frac{\partial}{\partial t}\left(\frac{\partial u_x}{\partial y}\right) + u_x\frac{\partial}{\partial x}\left(\frac{\partial u_x}{\partial y}\right) + u_y\frac{\partial}{\partial y}\left(\frac{\partial u_x}{\partial y}\right) = -\left(\frac{1}{\rho}\frac{\partial^2 p}{\partial x \partial y}\right) + \nu\left[\frac{\partial^2}{\partial x^2}\left(\frac{\partial u_x}{\partial y}\right)+ \frac{\partial^2}{\partial y^2}\left(\frac{\partial u_x}{\partial y} \right)\right]$$

    $$\frac{\partial}{\partial t}\left(\frac{\partial u_y}{\partial x}\right) + u_x\frac{\partial}{\partial x}\left(\frac{\partial u_y}{\partial x}\right) + u_y\frac{\partial}{\partial y}\left(\frac{\partial u_y}{\partial x}\right) = -\left(\frac{1}{\rho}\frac{\partial^2 p}{\partial x\partial y}\right) + \nu\left[\frac{\partial^2}{\partial x^2}\left(\frac{\partial u_y}{\partial x}\right) + \frac{\partial^2}{\partial y^2}\left(\frac{\partial u_y}{\partial x} \right)\right]$$

* Subtracting the second equation from the first equation:

    $$\frac{\partial}{\partial t}\left(\frac{\partial u_x}{\partial y} - \frac{\partial u_y}{\partial x}\right) + u_x \frac{\partial}{\partial x}\left(\frac{\partial u_x}{\partial y} - \frac{\partial u_y}{\partial x}\right) + u_y \frac{\partial}{\partial y}\left(\frac{\partial u_x}{\partial y} - \frac{\partial u_y}{\partial x}\right) = \nu\left[\frac{\partial^2}{\partial x^2}\left(\frac{\partial u_x}{\partial y} - \frac{\partial u_y}{\partial x}\right)+ \frac{\partial^2}{\partial y^2}\left(\frac{\partial u_x}{\partial y} - \frac{\partial u_y}{\partial x}\right)\right]$$

* Vorticity is defined as the curl of velocity field and for 2-D flows this is given by:

    $$\omega = \nabla \times \mathbf{u} = det \begin{vmatrix} \frac{\partial}{\partial x} && \frac{\partial}{\partial y}\\ u_x && u_y \end{vmatrix} = \left(\frac{\partial u_y}{\partial x} - \frac{\partial u_x}{\partial y}\right)$$
    
            
* Substituting vorticity into the previous equation. The term being substituted is essentially the negative of vorticity, but the resulting equation is positive once you divide through by -1:

    $$\frac{\partial \omega}{\partial t} + u_x \frac{\partial \omega}{\partial x} + u_y \frac{\partial \omega}{\partial y} = \nu\left(\frac{\partial^2 \omega}{\partial x^2}+ \frac{\partial^2 \omega}{\partial y^2}\right)$$

* Stream function definition:

    $$\frac{\partial \psi}{\partial x} = -u_y$$

    $$\frac{\partial \psi}{\partial y} = u_x$$

* There are three unknowns ($\omega$, $u_x$ and $u_y$) in the vorticity-velocity equation. The Poisson's equation for the stream function derived below can be used to specify both velocities:

    $$\frac{\partial u_x}{\partial y} - \frac{\partial u_y}{\partial x} = -\omega$$

    $$\frac{\partial}{\partial y}\left(\frac{\partial \psi}{\partial y}\right) + \frac{\partial}{\partial x}\left(\frac{\partial \psi}{\partial x}\right) = -\omega$$

    $$\frac{\partial^2 \psi}{\partial y^2} + \frac{\partial^2 \psi}{\partial x^2} = -\omega$$

* It can be shown that the stream function satisfies continuity, implying that the continuity equation needs not be solved when dealing in stream function:

    $$\frac{\partial}{\partial x}\left(\frac{\partial \psi}{\partial y}\right) = \frac{\partial u_x}{\partial x}$$

    $$- \frac{\partial}{\partial y}\left(\frac{\partial \psi}{\partial x}\right) = \frac{\partial u_y}{\partial y}$$

    $$\frac{\partial u_x}{\partial x} + \frac{\partial u_y}{\partial y} = \frac{\partial^2 \psi}{\partial x \partial y} - \frac{\partial^2 \psi}{\partial x \partial y} = 0$$

* Thus, we only need to solve:

    $$\frac{\partial \omega}{\partial t} + u_x \frac{\partial \omega}{\partial x} + u_y \frac{\partial \omega}{\partial y} = \nu\left(\frac{\partial^2 \omega}{\partial x^2}+ \frac{\partial^2 \omega}{\partial y^2}\right)$$

    $$\frac{\partial^2 \psi}{\partial y^2} + \frac{\partial^2 \psi}{\partial x^2} = -\omega$$

# PART 6: ADVECTION-DIFFUSION SCHEMES
****

* Compact form of steady advection-diffusion equation:

    $$\nabla \cdot \left(\rho u \phi\right) - \nabla \cdot \left(\Gamma\nabla \phi\right) = S_{\phi}$$

* Integrating over the control volume:

    $$\int_{cv} \left[\nabla \cdot \left(\rho u \phi\right) - \nabla \cdot \left(\Gamma\nabla \phi\right)\right] dv = \int_{cv} S_{\phi} dV$$

* Applying Gauss divergence theorem to the advection and diffusion term:

    $$\int_{cs} \mathbf{n} \cdot \left[\left(\rho u \phi\right) - \left(\Gamma\nabla \phi\right)\right] dS = \int_{cv} S_{\phi} dV$$

* For 1-D problem:

    $$\left[\left(\rho u \phi\right)_e - \left(\rho u \phi\right)_w \right] + \left[\Gamma_w \frac{\phi_P - \phi_W}{\Delta x} - \Gamma_e \frac{\phi_E - \phi_P}{\Delta x} \right] = S_{\phi} \Delta x$$

* The central difference scheme is used above to evaluate the gradient in the diffusion term. Below, various schemes will be compared for the evaluation of advection fluxes at the east $\left(\rho u \phi\right)_e$ and west faces $\left(\rho u \phi\right)_w$.

**Central Difference Scheme**

* The central difference scheme assumes a piece-wise linear profile between two neighbouring nodes. From incompressible 1-D continuity equation $\rho u$ is constant and if we also assume $\Gamma$ is constant:

    $$\rho u\left[\frac{\phi_P + \phi_E}{2} - \frac{\phi_P + \phi_W}{2}\right] + \Gamma\left[\frac{\phi_P - \phi_W}{\Delta x} - \frac{\phi_E - \phi_P}{\Delta x} \right] = S_{\phi} \Delta x$$

* Re-arranging the advection term, we see the central difference scheme is oblivious to the state of the node of interest (P). This can result in a non-physical solution especially if the field does not obey continuity at any point during the simulation:

    $$\rho u\left(\frac{\phi_E - \phi_W}{2}\right) + \frac{\Gamma}{\Delta x} \left[2\phi_P - \phi_W - \phi_E\right] = S_{\phi} \Delta x$$

* Multiplying through by $\frac{\Delta x}{\Gamma}$ and isolating $\phi_P$:

    $$\phi_P = -\frac{\rho u \Delta x}{2\Gamma}\left(\frac{\phi_E - \phi_W}{2}\right) + \left[\frac{\phi_W + \phi_E}{2} \right] + S_{\phi} \frac{\Delta x^2}{2\Gamma}$$

* Setting $S_\phi = 0$ and subtracting $\phi_W$ from both sides:

    $$\phi_P - \phi_W = -\frac{Pe_L}{2}\left(\frac{\phi_E - \phi_W}{2}\right) + \left[\frac{\phi_E - \phi_W}{2} \right]$$

* Dividing through by $\left(\phi_E - \phi_W\right)$:

    $$\frac{\phi_P - \phi_W}{\phi_E - \phi_W} = \frac{1}{2}\left[1 - \frac{Pe_L}{2}\right]$$
    
**Exact Scheme**

* Considering steady 1-D advection-diffusion problem in x, the governing equation is given by:

    $$\frac{d \left(\rho u \phi\right)}{dx} = \frac{d}{dx}\left(\Gamma \frac{d\phi}{dx}\right)$$

* Integrating both sides once:

    $$\int \rho u\frac{d\phi}{dx} dx = \int \frac{d}{dx}\left(\Gamma \frac{d\phi}{dx}\right) dx$$

    $$\frac{d \phi}{dx} = \frac{\rho u \phi}{\Gamma} - \frac{C_1}{\Gamma}$$

* Change of variable:

    $$\Phi = \frac{\rho u \phi}{\Gamma} - \frac{C_1}{\Gamma}$$

* Differentiating the new variable with respect to x:

    $$\frac{d\Phi}{dx} = \frac{\rho u}{\Gamma}\frac{d\phi}{dx}$$

    $$\frac{d\phi}{dx} = \frac{\Gamma}{\rho u}\frac{d\phi}{dx}$$

* With the above change of variable, we get:

    $$\frac{d\Phi}{dx} = \frac{\rho u \Phi}{\Gamma}$$

* Separating variables and integrating:

    $$\int \frac{d\Phi}{\Phi} = \int \frac{\rho u}{\Gamma} dx$$

    $$ln\left(\Phi\right) = \frac{\rho u x}{\Gamma} + C_3$$

    $$\Phi = C_2 e^{\left(\frac{\rho u x}{\Gamma}\right)}$$

* Subbing in  the definition for $\Phi$:

    $$\phi = \frac{C_2 \Gamma e^{\left(\frac{\rho u x}{\Gamma}\right)} + C_1}{\rho u}$$

* For east and west grid points:

    $$\phi_E = \frac{C_2 \Gamma e^{\left(\frac{\rho u x_E}{\Gamma}\right)} + C_1}{\rho u}$$

    $$\phi_W = \frac{C_2 \Gamma e^{\left(\frac{\rho u x_W}{\Gamma}\right)} + C_1}{\rho u}$$

* The dimensionless quantity $\frac{\phi - \phi_W}{\phi_E - \phi_W}$ can be obtained for the exact solution as follows:

    $$\frac{\phi - \phi_W}{\phi_E - \phi_W} = \frac{e^{\left(\frac{\rho u x}{\Gamma}\right)} - e^{\left(\frac{\rho u x_W}{\Gamma}\right)}}{e^{\left(\frac{\rho u x_E}{\Gamma}\right)} -  e^{\left(\frac{\rho u x_W}{\Gamma}\right)}}$$

    $$\frac{\phi - \phi_W}{\phi_E - \phi_W} = \frac{e^{\left(\frac{\rho u x_W}{\Gamma}\right)}\left[e^{\left(\frac{\rho u (x-x_W)}{\Gamma}\right)} - 1 \right]}{e^{\left(\frac{\rho u x_W}{\Gamma}\right)}\left[e^{\left(\frac{\rho u (x_E-X_W)}{\Gamma}\right)} -  1\right]}$$

    $$\frac{\phi - \phi_W}{\phi_E - \phi_W} = \frac{\left[e^{\left(\frac{\rho u (x-x_W)}{\Gamma}\right)} - 1 \right]}{\left[e^{\left(\frac{\rho u (x_E-X_W)}{\Gamma}\right)} -  1\right]}$$

* Recognizing that $x-x_W = \Delta x$ and $x_E-x_W = 2\Delta x$:

    $$\frac{\phi - \phi_W}{\phi_E - \phi_W} = \frac{\left[e^{\left(\frac{\rho u \Delta x}{\Gamma}\right)} - 1 \right]}{\left[e^{\left(2\frac{\rho u \Delta x}{\Gamma}\right)} -  1\right]}$$

    $$\frac{\phi - \phi_W}{\phi_E - \phi_W} = \frac{e^{Pe_L} - 1}{e^{\left(2Pe_L\right)} -  1}$$

* The value at the node of interest is represented as $\phi$ here. To be consistent with the central difference scheme, we replace it with $\phi_P$:

    $$\frac{\phi_P - \phi_W}{\phi_E - \phi_W} = \frac{e^{Pe_L} - 1}{e^{\left(2Pe_L\right)} -  1}$$

* It should be noted that the exact scheme is limited to steady 1-D advection-diffusion problems and its extension to multidimensional problems may be problematic.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Peclet numbers
Pe = np.linspace(-10,10,1000)

# central difference
dimless_phi_cd = 1.0/2.0*(1.0-Pe/2.0)

# exact solution
dimless_phi_exact = (np.exp(Pe) - 1.0)/(np.exp(2*Pe) - 1.0)

# compare solutions
plt.figure()
plt.plot(Pe,dimless_phi_cd,'b-',label=r'Central Difference')
plt.plot(Pe, dimless_phi_exact,'r-', label=r'Exact Solution')
plt.xlabel(r'$Pe_L$')
plt.ylabel(r'$\frac{\phi_P - \phi_W}{\phi_E - \phi_W}$')
plt.legend()
plt.show()
```

**Upwind Scheme**

* From the graphical comparison of central difference and exact scheme, we see that the central difference scheme is valid for a narrow range of Peclet numbers ($-2.5 < Pe_L < 2.5$). On the other hand, the exact scheme is valid for 1-D problems only. For multidimensional problems, a different scheme is required and one of the available options is the upwind scheme

* From above, the discretized steady 1-D advection-diffusion equation ($S_{\phi} = 0$) is:

    $$\left[\left(\rho u \phi\right)_e - \left(\rho u \phi\right)_w \right] + \left[\Gamma_w \frac{\phi_P - \phi_W}{\Delta x} - \Gamma_e \frac{\phi_E - \phi_P}{\Delta x} \right] = 0$$

* We need to evaluate the face fluxes from the discretized advection term. The upwind scheme uses the value of quantity of interest at the node upstream. 

    $$\phi_e = \begin{cases} \phi_P \text{ if } \left(\rho u\right)_e > 0 \\ \phi_E \text{ if } \left(\rho u\right)_e < 0\end{cases}$$

    $$\phi_w = \begin{cases} \phi_W \text{ if } \left(\rho u\right)_w > 0 \\ \phi_P \text{ if } \left(\rho u\right)_w < 0\end{cases}$$

* The above conditionals can be re-expressed as follows, with $F_e = \left(\rho u\right)_e$ and $F_w = \left(\rho u\right)_w$:

    $$F_e \phi_e = \phi_P \max \left(F_e, 0\right) - \phi_E \max \left(-F_e, 0\right)$$

    $$F_w \phi_w = \phi_W \max \left(F_w, 0\right) - \phi_P \max \left(-F_w, 0\right)$$

* Subbing above expressions into the discretized equation:

    $$\left[\phi_P \max \left(F_e, 0\right) - \phi_E \max \left(-F_e, 0\right)\right] -  \left[\phi_W \max \left(F_w, 0\right) - \phi_P \max \left(-F_w, 0\right)\right] + \left[\Gamma_w \frac{\phi_P - \phi_W}{\Delta x} - \Gamma_e \frac{\phi_E - \phi_P}{\Delta x} \right] = 0$$

* From continuity $\rho u$ is constant and assuming constant $\Gamma$:

    $$\left[\phi_P \max \left(F, 0\right) - \phi_E \max \left(-F, 0\right)\right] -  \left[\phi_W \max \left(F, 0\right) - \phi_P \max \left(-F, 0\right)\right] + \frac{\Gamma}{\Delta x}\left[2\phi_P - \phi_W - \phi_E \right] = 0$$

* Multiplying through $\frac{\Delta x}{\Gamma}$:

    $$\left[\phi_P \max \left(Pe_L, 0\right) - \phi_E \max \left(-Pe_L, 0\right)\right] -  \left[\phi_W \max \left(Pe_L, 0\right) - \phi_P \max \left(-Pe_L, 0\right)\right] + 2\phi_P - \phi_W - \phi_E = 0$$

* Re-arranging:

    $$\left[2 + \max \left(Pe_L, 0\right) + \max \left(-Pe_L, 0\right) \right] \phi_P = \left[1 + \max \left(-Pe_L, 0\right)\right]\phi_E + \left[1 + \max \left(Pe_L, 0\right)\right]\phi_W$$

    $$a_P \phi_P = a_E \phi_E + a_W \phi_W$$
    
    $$a_P = a_E + a_W$$

* Expressing $\phi_P$ in terms of $\phi_E$, $\phi_W$, and coefficients ($a_E$, $a_W$, and $a_P$)

    $$\phi_P = \frac{a_E \phi_E + a_W\phi_W}{a_P}$$

    $$\phi_P = \frac{a_E \phi_E + a_W\phi_W}{a_E + a_W}$$

* Subtracting $\phi_W$ from both sides of the equation:

    $$\phi_P - \phi_W = \frac{a_E \phi_E + a_W\phi_W}{a_E + a_W} - \phi_W$$

    $$\phi_P - \phi_W = \frac{a_E \phi_E + a_W\phi_W - a_E \phi_W - a_W\phi_W}{a_E + a_W}$$

    $$\phi_P - \phi_W = \frac{a_E \left(\phi_E - \phi_W\right)}{a_E + a_W}$$

* Dividing through by $\phi_E - \phi_W$:

    $$\frac{\phi_P - \phi_W}{\phi_E - \phi_W} = \frac{a_E}{a_E + a_W}$$

* Plugging in the expressions for $a_E$ and $a_W$:

    $$\frac{\phi_P - \phi_W}{\phi_E - \phi_W} = \frac{1 + \max \left(-Pe_L, 0\right)}{2 + \max \left(Pe_L, 0\right) + \max \left(-Pe_L, 0\right)}$$

```python
# upwind scheme solution
dimless_phi_upwind = (1.0 + np.maximum(-Pe,0))/(2.0 + np.maximum(Pe,0) + np.maximum(-Pe,0))

# compare to other two schemes
plt.figure()
plt.plot(Pe,dimless_phi_cd,'b-',label=r'Central Difference')
plt.plot(Pe, dimless_phi_exact,'r-', label=r'Exact Solution')
plt.plot(Pe, dimless_phi_upwind,'g-',label=r'Upwind Scheme')
plt.xlabel(r'$Pe_L$')
plt.ylabel(r'$\frac{\phi_P - \phi_W}{\phi_E - \phi_W}$')
plt.legend()
plt.show()
```
