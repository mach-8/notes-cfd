***
# GENERAL GOVERNING EQUATIONS
***


### CONTINUITY EQUATION 

* Mass conservation requires:
    
    $$\text{Accumulation } + [\text{Out} - \text{In}] = 0$$
    
* Accumulation:
    
    $$\frac{\left[\rho|_{t+\Delta t} - \rho|_t\right]\Delta x \Delta y \Delta z}{\Delta t}$$
    
* Net outflow through x:
    
    $$\left[\rho u_x|_{x+\Delta x} - \rho u_x|_x\right]\Delta y \Delta z$$
    
* Net outflow through y:

    $$\left[\rho u_y|_{y+\Delta y} - \rho u_y|_y\right]\Delta x \Delta z$$
    
* Net outflow through z:

    $$\left[\rho u_z|_{z+\Delta z} - \rho u_z|_z\right]\Delta x \Delta y$$
    
* Combining all terms:
    
    $$\frac{\left[\rho|_{t+\Delta t} - \rho|_t\right]\Delta x \Delta y \Delta z}{\Delta t} + \left[\rho u_x|_{x+\Delta x} - \rho u_x|_x\right]\Delta y \Delta z + \left[\rho u_y|_{y+\Delta y} - \rho u_y|_y\right]\Delta x \Delta z + \left[\rho u_z|_{z+\Delta z} - \rho u_z|_z\right]\Delta x \Delta y = 0$$

* Dividing through by the volume of the fluid element, $\Delta x \Delta y \Delta z$:

    $$\frac{\left[\rho|_{t+\Delta t} - \rho|_t\right]}{\Delta t} + \frac{\left[\rho u_x|_{x+\Delta x} - \rho u_x|_x\right]}{\Delta x} + \frac{\left[\rho u_y|_{y+\Delta y} - \rho u_y|_y\right]}{\Delta y} + \frac{\left[\rho u_z|_{z+\Delta z} - \rho u_z|_z\right]}{\Delta z} = 0$$
    
* In the limits of $\Delta t$, $\Delta x$, $\Delta y$, and $\Delta z$ approaching zero:

    $$\frac{\partial \rho}{\partial t} + \frac{\partial \left(\rho u_x\right)}{\partial x} + \frac{\partial \left(\rho u_y\right)}{\partial y} + \frac{\partial \left(\rho u_z\right)}{\partial z} = 0$$

### MOMENTUM BALANCE EQUATIONS

* The rate of change of a conserved quantity $\phi\left(t, x, y, z\right)$ following a fluid particle is given by the total/material/substantial derivative of the quantity.

    $$\frac{D\phi}{Dt} = \frac{\partial \phi}{\partial t} + \frac{\partial \phi}{\partial x} \frac{\partial x}{\partial t} + \frac{\partial \phi}{\partial y}\frac{\partial y}{\partial t} + \frac{\partial \phi}{\partial z}\frac{\partial z}{\partial t}$$
    
* Recognizing that $\frac{\partial x}{\partial t} = u_x$, $\frac{\partial y}{\partial t} = u_y$, and $\frac{\partial z}{\partial t} = u_z$:
    
    $$\frac{D\phi}{Dt} = \frac{\partial \phi}{\partial t} + u_x \frac{\partial \phi}{\partial x} + u_y \frac{\partial \phi}{\partial y} + u_z \frac{\partial \phi}{\partial z}$$

* The above equation describes the rate of change per unit mass using Lagrangian approach - the rate of change per unit volume (Eulerian approach) can be obtained by multiplying the prior equation by the density:

    $$\rho \frac{D\phi}{Dt} = \rho \left[ \frac{\partial \phi}{\partial t} + u_x \frac{\partial \phi}{\partial x} + u_y \frac{\partial \phi}{\partial y} + u_z \frac{\partial \phi}{\partial z} \right]$$
    
* It can alternatively be derived by considering a fluid element fixed in space. The total rate of change for the fluid element comprises of time rate of change and advection from the surroundings through the boundaries of the fluid element.

    $$ \text{Total Rate of Change} = \text{Time Rate of Change} + \text{Advection from Surroundings}$$
    
    $$\rho \Delta x \Delta y \Delta z \frac{D\phi}{Dt} = \frac{\left[\rho \phi|_{t + \Delta t} - \rho \phi|_t\right]\Delta x \Delta y \Delta z}{\Delta t} + \left[\rho u_x \phi|_{x+\Delta x} - \rho u_x \phi|_x\right]\Delta y \Delta z + \left[\rho u_y \phi|_{y + \Delta y} - \rho u_y \phi|_y\right] \Delta x \Delta z + \left[\rho u_z \phi|_{z + \Delta z} - \rho u_z \phi|_z\right] \Delta x \Delta y$$
    
     $$\rho \frac{D\phi}{Dt} = \frac{\left[\rho \phi|_{t + \Delta t} - \rho \phi|_t\right]}{\Delta t} + \frac{\left[\rho u_x \phi|_{x+\Delta x} - \rho u_x \phi|_x\right]}{\Delta x} + \frac{\left[\rho u_y \phi|_{y + \Delta y} - \rho u_y \phi|_y\right]}{\Delta y} + \frac{\left[\rho u_z \phi|_{z + \Delta z} - \rho u_z \phi|_z\right]}{\Delta z}$$
    
    $$\rho \frac{D\phi}{Dt} = \frac{\partial \left(\rho \phi\right)}{\partial t} + \frac{\partial \left(\rho u_x \phi\right)}{\partial x} + \frac{\partial \left(\rho u_y \phi\right)}{\partial y} + \frac{\partial \left(\rho u_z \phi\right)}{\partial z}$$
    
     $$\rho \frac{D\phi}{Dt} = \rho \left[\frac{\partial \phi}{\partial t} + u_x\frac{\partial \phi}{\partial x} + u_y\frac{\partial \phi}{\partial y} + u_z\frac{\partial \phi}{\partial z}\right] + \phi\left[\frac{\partial \rho}{\partial t} + \frac{\partial \left(\rho u_x\right)}{\partial x} + \frac{\partial \left(\rho u_y\right)}{\partial y} + \frac{\partial \left(\rho u_z\right)}{\partial z}\right]$$
     
    $$\rho \frac{D\phi}{Dt} = \rho \left[\frac{\partial \phi}{\partial t} + u_x\frac{\partial \phi}{\partial x} + u_y\frac{\partial \phi}{\partial y} + u_z\frac{\partial \phi}{\partial z}\right]$$
     
* From Newton's second law of motion, the rate of change of momentum (conserved quantity) is equal to the sum of forces acting on the fluid element:
    
    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = \sum F_x$$
    
    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = \sum F_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = \sum F_z$$

* Forces (per unit volume) under consideration:
    
    1. Pressure forces: 

        * x-component: $-\frac{\partial p}{\partial x}$

        * y-component: $-\frac{\partial p}{\partial y}$

        * z-component: $-\frac{\partial p}{\partial z}$
    
    2. Viscous shear forces (first subscript is the cordinate perpendicular to the force and second subscript is the direction in which the force is acting): 

        * x-components: $\frac{\partial \tau_{xx}}{\partial x}$, $\frac{\partial \tau_{yx}}{\partial x} $, $\frac{\partial \tau_{zx}}{\partial x}$
        
        * y-components:   $\frac{\partial \tau_{xy}}{\partial y}$, $\frac{\partial \tau_{yy}}{\partial y}$, $\frac{\partial \tau_{zy}}{\partial y}$
        
        * z-components:  $\frac{\partial \tau_{xz}}{\partial z}$, $\frac{\partial \tau_{yz}}{\partial z} $, $\frac{\partial \tau_{zz}}{\partial z}$

    3. Gravity: 
    
        * x-component: $\rho g_x$
        
        * y-component: $\rho g_y$ 
        
        * z-component: $\rho g_z$

    4. Other forces: 
        
        * x-component: $S_x$ 
        
        * y-component: $S_y$ 
        
        * z-component: $S_z$

* Plugging in the forces:

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \frac{\partial \tau_{xx}}{\partial x} + \frac{\partial \tau_{yx}}{\partial x} + \frac{\partial \tau_{zx}}{\partial x} + \rho g_x + S_x$$
    
    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \frac{\partial \tau_{xy}}{\partial y} + \frac{\partial \tau_{yy}}{\partial y} + \frac{\partial \tau_{zy}}{\partial y} + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \frac{\partial \tau_{xz}}{\partial z} + \frac{\partial \tau_{yz}}{\partial z} + \frac{\partial \tau_{zz}}{\partial z} + \rho g_z + S_z$$


***
# GOVERNING EQUATION FOR NEWTONIAN FLUIDS
***

### CONTINUITY EQUATION

* The continuity equation remains unchanged as shown above:

    $$\frac{\partial \rho}{\partial t} + \frac{\partial \left(\rho u_x\right)}{\partial x} + \frac{\partial \left(\rho u_y\right)}{\partial y} + \frac{\partial \left(\rho u_z\right)}{\partial z} = 0$$
    

### MOMENTUM BALANCE EQUATIONS

* For Newtonian fluids, we can use Stokes' relations for viscous stresses:

    - Linear transformations: 
        
        $$\tau_{xx} = 2\mu \frac{\partial u_x}{\partial x} + \lambda \nabla \cdot \mathbf{u}$$
        
        $$\tau_{yy} = 2\mu \frac{\partial u_y}{\partial y} + \lambda \nabla \cdot \mathbf{u}$$
        
        $$\tau_{zz} = 2\mu \frac{\partial u_z}{\partial z} + \lambda \nabla \cdot \mathbf{u}$$

    - Volumetric transformations: 
        
        $$\tau_{xy} = \tau_{yx} = \mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)$$
        
        $$\tau_{xz} = \tau_{zx} = \mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)$$
        
        $$\tau_{yz} = \tau_{zy} = \mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)$$

* Plugging Stokes' relations into the general momentum balance equations:
    
    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \frac{\partial}{\partial x}\left[2\mu \frac{\partial u_x}{\partial x} + \lambda \nabla \cdot \mathbf{u} \right] + \frac{\partial}{\partial y}\left[\mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)\right]  + \frac{\partial}{\partial z}\left[\mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\right] + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \frac{\partial}{\partial x}\left[\mu \left(\frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)\right] + \frac{\partial}{\partial y}\left[2\mu \frac{\partial u_y}{\partial y} + \lambda \nabla \cdot \mathbf{u} \right] + \frac{\partial}{\partial z}\left[\mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\right] + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \frac{\partial}{\partial x}\left[\mu \left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)\right] + \frac{\partial}{\partial y}\left[\mu \left(\frac{\partial u_z}{\partial y} + \frac{\partial u_y}{\partial z}\right)\right] + \frac{\partial}{\partial z}\left[2\mu \frac{\partial u_z}{\partial z} + \lambda \nabla \cdot \mathbf{u} \right] + \rho g_z + S_z$$

* Re-arranging:

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \mu \frac{\partial}{\partial x}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \mu \frac{\partial}{\partial y}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial y}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \mu \frac{\partial}{\partial z}\left(\frac{\partial u_x}{\partial x}+ \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_z + S_z$$

* In more compact form:

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \mu \frac{\partial}{\partial x}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \mu \frac{\partial}{\partial y}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial y}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \mu \frac{\partial}{\partial z}\left(\nabla \cdot \mathbf{u}\right)  + \frac{\partial}{\partial x}\left(\lambda \nabla \cdot \mathbf{u}\right) + \rho g_z + S_z$$

# GOVERNING EQUATIONS FOR INCOMPRESSIBLE FLOWS
***

### CONTINUITY EQUATION

* For incompressible flows, the density is constant, so we can set $\frac{\partial \rho}{\partial t} = 0$:

    $$\frac{\partial u_x}{\partial x} + \frac{\partial u_y}{\partial y} + \frac{\partial u_z}{\partial z} = 0$$

* In more compact form:

    $$\nabla \cdot \mathbf{u} = 0$$

### MOMENTUM BALANCE EQUATIONS

* We can use the continuity equation for incompressible flows shown above to simplify the momentum balance equations for Newtonian fluids by setting $\nabla \cdot \mathbf{u} = 0$. This should result in the following:

    $$\rho\left[\frac{\partial u_x}{\partial t} + u_x\frac{\partial u_x}{\partial x} + u_y\frac{\partial u_x}{\partial y} + u_z\frac{\partial u_x}{\partial z}\right] = -\frac{\partial p}{\partial x} + \mu \left(\frac{\partial ^2 u_x}{\partial x^2} + \frac{\partial ^2 u_x}{\partial y^2} + \frac{\partial ^2 u_x}{\partial z^2}\right) + \rho g_x + S_x$$

    $$\rho\left[\frac{\partial u_y}{\partial t} + u_x\frac{\partial u_y}{\partial x} + u_y\frac{\partial u_y}{\partial y} + u_z\frac{\partial u_y}{\partial z}\right] = -\frac{\partial p}{\partial y} + \mu \left(\frac{\partial ^2 u_y}{\partial x^2} + \frac{\partial ^2 u_y}{\partial y^2} + \frac{\partial ^2 u_y}{\partial z^2}\right) + \rho g_y + S_y$$

    $$\rho\left[\frac{\partial u_z}{\partial t} + u_x\frac{\partial u_z}{\partial x} + u_y\frac{\partial u_z}{\partial y} + u_z\frac{\partial u_z}{\partial z}\right] = -\frac{\partial p}{\partial z} + \mu \left(\frac{\partial ^2 u_z}{\partial x^2} + \frac{\partial ^2 u_z}{\partial y^2} + \frac{\partial ^2 u_z}{\partial z^2}\right) + \rho g_z + S_z$$
